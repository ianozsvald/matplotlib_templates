```python
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
values = scipy.random.normal(size=10)
```

```python
fig, ax = plt.subplots(figsize=(8,6)); # returns 1 or more axes 
plt.plot(values, axes=ax);
plt.title("Some random data")
```

```python
fig, ax = plt.subplots(figsize=(8,6)); # returns 1 or more axes 
plt.plot(values, axes=ax);
plt.title("Some random data")

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
```

```python
fig, ax = plt.subplots(figsize=(8,6)); # returns 1 or more axes 
plt.plot(values, axes=ax);
plt.title("Some random data")

# props
# https://stackoverflow.com/questions/14349055/making-matplotlib-graphs-look-like-r-by-default

# remove right/top border
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# offset 
ax.spines['bottom'].set_position(('axes', -0.05))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('axes', -0.05))

plt.grid(b=True, which='major', color='gray', linestyle='-')
#plt.grid(b=True, which='minor', color='r', linestyle='--')

plt.locator_params(nbins=10)
# http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.locator_params
# defaults to both, can be for x or y only

fig.tight_layout()
```
